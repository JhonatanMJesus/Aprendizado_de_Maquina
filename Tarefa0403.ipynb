### Ex. 1

Crie uma classe Aluno que armazene o nome e uma lista de notas. Implemente um
método para calcular a média e outro para verificar se o aluno foi aprovado (média >= 6).

```python
class Aluno:
    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas

    def calcular_media(self):
        media = sum(self.notas) / len(self.notas)
        return media
    
    def classificar(self):
        media = self.calcular_media()
        if media >= 6:
            return "Aprovado"
        else:
            return "Reprovado"
        
    # Na prática:
aluno1 = Aluno("João", [6, 8, 5])

aluno2 = Aluno("Joaquim", [4, 7])

print(f"Aluno: {aluno1.nome}, Média: {aluno1.calcular_media()}, Classificação: {aluno1.classificar()}")
print(f"Aluno: {aluno2.nome}, Média: {aluno2.calcular_media()}, Classificação: {aluno2.classificar()}")
```

### Ex. 2

Crie uma classe Conversor que possua um atributo privado taxa_dolar. Implemente 
métodos para atualizar a taxa (com validação para não aceitar valores negativos) e 
para converter um valor de Real para Dólar.

```python
class Conversor:
    def __init__(self):
        self.__taxa_dolar = 1.0

    def atualizar_taxa(self, nova_taxa):
        if nova_taxa > 0:
            self.__taxa_dolar = nova_taxa
    
    def converter_para_dolar(self, valor_reais):
        valor_dolar = valor_reais / self.__taxa_dolar
        return valor_dolar
    
conversor = Conversor()
conversor.atualizar_taxa(5.39)

valor_em_reais = 100
valor_em_dolar = conversor.converter_para_dolar(valor_em_reais)

print(f"Valor em dólares: ${valor_em_dolar:.2f}")
```

### Ex.3

Crie as classes Livro e Biblioteca. A classe Biblioteca deve conter uma lista de 
objetos do tipo Livro. Implemente métodos para adicionar livros e listar todos os 
títulos disponíveis.

```python
class Livro:
    def __init__(self, titulo, isbn):
        self.titulo = titulo
        self.isbn = isbn

class Biblioteca:
    def __init__(self, nome):
        self.nome = nome
        self.livros = []

    def adicionar_livro(self, livro):
        self.livros.append(livro)

    def listar_livros(self):
        for livro in self.livros:
            print(f"Livro: {livro.titulo}, ISBN: {livro.isbn}")

livro1 = Livro('João e Maria', 1254)
livro2 = Livro('Jogos Vorazes', 1562)

biblioteca1 = Biblioteca('Biblioteca da Fatec')
biblioteca1.adicionar_livro(livro1)
biblioteca1.adicionar_livro(livro2)
biblioteca1.listar_livros()
```

### Ex. 4

Crie uma classe base Veiculo com atributos como marca e modelo. Crie
subclasses Carro (com número de portas) e Moto (com cilindradas). Ambas devem
herdar as características da classe pai.

```python
class Veiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo

class Carro(Veiculo):
    def __init__(self, marca, modelo, portas):
        super().__init__(marca, modelo)
        self.portas = portas

class Moto(Veiculo):
    def __init__(self, marca, modelo, cilindradas):
        super().__init__(marca, modelo)
        self.cilindradas = cilindradas

carro1 = Carro("Chevrolet", "Ônix", 4)
moto1 = Moto("Honda", "CB300R", 291)
print(f"Carro \nMarca: {carro1.marca} \nModelo: {carro1.modelo} \nQtd. Portas: {carro1.portas}\n")
print(f"Moto \nMarca: {moto1.marca} \nModelo: {moto1.modelo} \nCilindradas: {moto1.cilindradas}")
```

### Ex. 6

No contexto de IoT, crie uma classe Sensor que armazene ID e tipo. Crie um script
que permita ao usuário cadastrar vários sensores em um dicionário, onde a chave
é o ID e o valor é o objeto Sensor.

```python
class Sensor:
    def __init__(self, id, tipo):
        self.id = id
        self.tipo = tipo

sensores = {}

while True:
    print("\n--- Página inicial ---")
    print("1 - Cadastrar Sensor")
    print("2 - Listar Sensores")
    print("3 - Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        id = input("Digite o ID do sensor: ")
        if id in sensores:
            print("Já existe um sensor com esse ID, tente novamente.")
        else:
            tipo = input("Digite o tipo de sensor: ")
            sensores[id] = Sensor(id, tipo)
            print(f"Sensor '{id} - {tipo}' cadastrado com sucesso")

    elif opcao == "2":
        if not sensores:
            print("Nenhum sensor cadastrado!")
        else:
            print("\n--- Sensores cadastrado ---")
            for chave, sensor in sensores.items():
                print(f"ID: {sensor.id} | Tipo: {sensor.tipo}")

    elif opcao == "3":
        print("Encerrando...")
        break

    else:
        print("\nOpção Inválida.")
```

### Ex. 7

Crie uma classe Usuario com login e senha. Implemente um método que valide se
a senha possui pelo menos 8 caracteres e um caractere especial. Se não atender,
o sistema deve solicitar uma nova senha

```python
import re

class Usuario:
    def __init__(self, login, senha):
        self.login = login
        self.senha = senha

    def validar_senha(self):
        while True:
            char_especiais = re.search(r'[!@#$%^&*(),.?":{}|<>]', self.senha)

            if len(self.senha) < 8 and not char_especiais:
                print(f"{self.login}, sua senha deve ter pelo menos 8 dígitos e 1 caractere especial")
            elif len(self.senha) < 8:
                print(f"{self.login}, sua senha tem {len(self.senha)} caracteres, mínimo: 8")
            elif not char_especiais:
                print(f"{self.login}, sua senha deve ter pelo menos 1 caractere especial")
            else:
                print(f"{self.login}, sua senha atende aos requisitos mínimos!")
                break

            self.senha = input("Digite sua nova senha: ")

usuario1 = Usuario("Jhonatan", "senha@123")
usuario2 = Usuario("Kieran", "sen123")

usuario1.validar_senha()
usuario2.validar_senha()
```

### Ex. 9

Crie uma classe Funcionario que possua um método registrar_entrada(). Utilize o 
módulo datetime para capturar o horário exato em que o método foi chamado e 
armazene isso em uma lista de logs dentro do objeto

```python
import datetime

class Funcionario:
    def __init__(self, nome):
        self.nome = nome
        self.logs = []

    def registrar_entrada(self):
        date_time = datetime.datetime.now()
        self.logs.append(date_time)
        print(f"{self.nome}, horário: {date_time} registrado!")

funcionario1 = Funcionario("Jhonatan")
funcionario1.registrar_entrada()
funcionario1.registrar_entrada()
funcionario1.registrar_entrada()
funcionario1.registrar_entrada()

for log in funcionario1.logs:
    print(f"Hora registrada: {log.strftime('%d/%m/%Y %H:%M:%S')}")
```

### Ex. 10

Crie  uma  classe  Calculadora  que  contenha  apenas  métodos  estáticos  para 
operações como soma, subtração e raiz quadrada. O objetivo é usar a classe como 
um "namespace" de funções utilitárias sem precisar instanciá-la.

```python
import math

class Calculadora:

    @staticmethod
    def soma(num1, num2):
        return num1 + num2
    
    @staticmethod
    def subtracao(num1, num2):
        return num1 - num2
    
    @staticmethod
    def raiz_quadrada(num):
        if num < 0:
            return "Não existe raiz de número negativo"
        return math.sqrt(num)

print(f"Soma: {Calculadora.soma(5, 8)}")
print(f"Subtração: {Calculadora.subtracao(10, 8)}")
print(f"Raiz quadrada: {Calculadora.raiz_quadrada(121)}")
```
